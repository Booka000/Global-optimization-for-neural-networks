```python
import numpy as np
import tensorflow as tf
import import_ipynb
from NN import get_dataset, get_model, get_shape, set_shape, get_con_mat, LossHistory
import matplotlib.pyplot as plt
import json
import os
from sklearn.metrics import confusion_matrix
```

```python
model = get_model()
```

```python
loss_history = LossHistory()
```
